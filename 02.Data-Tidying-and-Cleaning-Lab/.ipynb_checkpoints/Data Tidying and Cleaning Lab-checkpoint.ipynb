### Problem 8. Coffee color by country and continent (1 point)
Create a table which shows how many coffees of each color are there in every country. Leave the missing values as they are.

**Note:** If you ask me, countries should be in rows, I prefer long tables much better than wide ones.

Now do the same for continents. You know what continent each country is located in.

```python
country_by_color = coffee_set_year_tidy.groupby(["country_of_origin", "color"]).agg(count = pd.NamedAgg(column = "color", aggfunc = "count"))
country_by_color = country_by_color.pivot_table(index = ["country_of_origin"], columns = "color", values = "count")
country_by_color = country_by_color.rename_axis(index=None, columns=None)
country_by_color.columns = country_by_color.columns.str.lower()
country_by_color.index =  country_by_color.index.str.lower()
country_by_color.to_csv("country_by_color")
```

```python
country_by_color = country_by_color.rename_axis('country').reset_index()
```

```python
country_by_color
```

```python
continents_set = pd.read_csv("data/countries_continents.csv")
continents_set.Continent
```

```python
missing_countries = [x for x in coffee_set_year_tidy.country_of_origin 
                     if (continents_set.Country.str.lower().str.contains(str(x).lower()).any() == False)]
missing_countries = np.unique(missing_countries)
missing_countries
```

```python
table_add = pd.DataFrame({'Continent':["Africa", "North America", "South America", "Asia", "Africa", "North America"], "Country": missing_countries[:-1]})
```

```python
table_add
```

```python
continents_set = pd.concat([continents_set,table_add],ignore_index=True)
```

```python
def continet_country(country):
    if continents_set.Country.str.lower().str.contains(country.lower()).any():
        return continents_set[continents_set.Country.str.lower().str.contains(country.lower())].iloc[0]["Continent"]#continents_set[continents_set.Country.str.lower() == "country"]["Continent"]
```

```python
continent_by_color = country_by_color
continent_by_color["continent"] = continent_by_color["country"].apply(continet_country)
```

```python
continent_by_color
```

```python
continent_by_color_grouped = continent_by_color.groupby("continent")[continent_by_color.columns[1:-1]].sum()#.apply(sum, axis = 0)
```

```python
continent_by_color_grouped = continent_by_color_grouped.reset_index()
continent_by_color_grouped.to_csv("continets_by_color")
```
