```python
precent_mean_ratio = (you_tube_views.youtube_likes/you_tube_views.youtube_views)
precent_mean_ratio
```

### Problem 8. TikTok stuff (2 points)
The most popular songs on TikTok released every year show... interesting behavior. Which years peaked the most TikTok views? Show an appropriate chart. Can you explain this behavior? For a bit of context, TikTok was created in 2016.

Now, how much popular is the most popular song for each release year, than the mean popularity? Analyze the results.

In both parts, it would be helpful to see the actual songs.

```python
spotify_songs.columns
```

```python
spotify_songs_titok = spotify_songs[["release_date",'tiktok_views', "track"]]
```

```python
spotify_songs_titok
```

```python
spotify_songs_titok.tiktok_views = spotify_songs_titok.tiktok_views.apply(lambda x: (float("{:e}".format(x))))
```

```python
spotify_songs_titok.release_date = spotify_songs_titok.release_date.dt.year.astype(int)
```

```python
spotify_songs_titok.dtypes
```

```python
views_per_year = spotify_songs_titok.groupby("release_date")["tiktok_views"].sum()
views_per_year
```

```python
plt.figure(figsize=(12, 5))
plt.bar(views_per_year.index, views_per_year)

plt.xticks(range(1987,2025, 2))

plt.xlabel("Year track release")
plt.ylabel("TikTok views")

plt.title("TikTok Views/Year")
plt.show()
```

```python
spotify_songs_titok
```

```python
index_max = spotify_songs_titok.groupby(["release_date"])["tiktok_views"].idxmax()
max_view_songs_year_tiktok = spotify_songs_titok.loc[index_max].reset_index(drop = True)
```

```python
max_view_songs_year_tiktok
```

```python
mean_tiktok = spotify_songs_titok.groupby(["release_date"])["tiktok_views"].mean()
mean_tiktok = mean_tiktok.reset_index()
```

```python
max_view_songs_year_tiktok["tiktok_dist"] = (max_view_songs_year_tiktok["tiktok_views"] - mean_tiktok.tiktok_views)
max_view_songs_year_tiktok
```

```python
plt.figure(figsize=(12, 7))

plt.bar(max_view_songs_year_tiktok.release_date, max_view_songs_year_tiktok.tiktok_dist, label = "most famous song view distribution")
plt.bar(max_view_songs_year_tiktok.release_date, max_view_songs_year_tiktok.tiktok_views, alpha= 0.5, label = "most famous song view")


plt.xlabel("Year")
plt.ylabel("Song TikTok famous")

plt.legend()
plt.title("Most famous song as per TikTok view counts and distribution value per Year")
plt.show()
```

#### Top year with max TikTok views are tracks released in 2022. 2023 result are middle of results from 2019 and 2020. Data for 2024 of course is not completed, because we are still in 2024. We can see that songs from 2005 and 2011 have a good count of interest in Tiktok.

#### Most famous song from 2005 has a big difference from the mean count of views in that year. There are very few or just 1 songs from the period of 1985 - 2000. Songs from the rest years has a similar views number or not enough count of songs to obtain better result.
