```python
def min_max_normalization(x, min = 0, max = 1):
    return (x - min) / (np.max(x) - np.min(x))
```

### Image Convolution

Valid padding - also called 'no padding,' doesn't add any extra pixels to the input image. Causing the output feature map to shrink compared to the input.  

Same padding - ensures the output has the same size as the input

scipy.signal.convolve2d

```python
conv_filter = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 0, 1]
])
```

```python
result = convolve2d(cat_image[:, :, 0], conv_filter, mode = "same")
```

```python
result.min(), result.max()
```

```python
imshow(result, cmap = "gray")
```
